```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
sample = np.random.randn(100)
```

```python
ecdf = np.cumsum(sample)
```

```python
ecdf[:10]
```

```python
sample.max()
```

```python
hist, bins = np.histogram(sample, bins=100)
```

```python
hist[-10:]
```

```python
fig = plt.figure(figsize=[15,10])
plt.plot(bins[:-1], np.cumsum(hist)/hist.sum())
plt.plot(bins[:-1], norm.cdf(bins[:-1]))
plt.show()
```
