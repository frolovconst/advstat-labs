```python
data = np.loadtxt('ds-boot-2.csv', delimiter='\t', skiprows=1)
B=5000
```

## Bootstrap estimation of 5 datset labels

The dataset is resampled in accordance with Bootstrap method B=5000 times. The results obtained demonstrate that the labels possess rather low variance.

```python
print('No of bootstrap samples: {}\n'.format(B))
print('Medians:')
for i in range(1,6):
    print('{}.  Median of y{}:'.format(i,i))
    bstrapped_param(np.median,data[:,i-6],B)
    
print('\nMeans:')
for i in range(1,6):
    print('{}.  Mean of y{}:'.format(i,i))
    bstrapped_param(np.mean,data[:,i-6],B)
```

## Linear regression parameters bootstrapping

The bootstrap estimation of the parameters was carried out in the following way. At first, the dataset was resampled uniformly B=5000 times. Each subsample was then fitted with linear equation. Thus B x 29 parameters of each of 5 targets were obtained. The variance of the parameters was then investigated, which demonstrated that all the parameters significantly vary across the subsamples (variance typically exceeding mean value) meaning that such a model cannot be used for making relaible predictions on such data. More precisely, overfitting cannot be avoided in this case.

```python
coefs = train_B_regressions(data, B)
analyze_coefs(coefs,B)
```
